# Exponential Approach to Terminal Velocity

This notebook simulates one dimensional motion under linear drag, showing how the velocity approaches a terminal value exponentially over time using a numerical time stepping method.


## Import Required Libraries


```python
import numpy as np
import matplotlib.pyplot as plt
```

## Model Parameters

Velocity and time are expressed in reduced units. The terminal velocity and characteristic time scale are set to unity for simplicity.


```python
v_ter = 1.0     # terminal speed magnitude
tau = 1.0       # characteristic time
dt = 0.01       # time step
t_max = 6.0     # maximum time
```

## Time Array


```python
t = np.arange(0, t_max + dt, dt)
```

## Velocity Array and Initial Condition


```python
v = np.zeros_like(t)
v[0] = -5 * v_ter
```

## Time Stepping Loop (Euler Method)

The equation of motion is  
$\displaystyle \frac{dv}{dt} = -\frac{v + v_{\text{ter}}}{\tau}$

```python
for i in range(1, len(t)):
    dvdt = (-v_ter - v[i-1]) / tau
    v[i] = v[i-1] + dvdt * dt
```

## Velocity vs Time Plot


```python
plt.figure(figsize=(8, 5))
plt.plot(t, v, 'o-', markersize=3, label="Numerical", color='blue')

plt.axhline(-v_ter, color='red', linestyle='--', label=r'$v_{\mathrm{ter}} = -1$')

plt.title('Exponential Approach to Terminal Velocity (Loop Simulation)')
plt.xlabel(r'$t/\tau$')
plt.ylabel(r'$v(t)$')
plt.legend()
plt.grid(True)
plt.show()
```
