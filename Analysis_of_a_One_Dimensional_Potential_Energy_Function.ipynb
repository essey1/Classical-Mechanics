# Analysis of a One Dimensional Potential Energy Function

This notebook analyzes a given potential energy function by plotting it, finding equilibrium points, determining their stability, and identifying turning points for a specified total energy.


## Import Required Libraries


```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, diff, Eq, solve
```

## Define Constants

W and d are positive constants that set the depth and spatial scale of the potential.


```python
W = 8.0   # positive constant
d = 5.0   # positive constant
```

## Part (a): Plot of the Potential U(x)


```python
x_vals = np.linspace(-10*d, 10*d, 1000)

U_vals = -W * (d**2 * (x_vals**2 + 4*d**2)) / (x_vals**4 + 128*d**4)
```

### Potential Energy Plot


```python
plt.plot(x_vals, U_vals)
plt.axhline(0, color='black', linewidth=0.5)
plt.xlabel('x')
plt.ylabel('U(x)')
plt.title('Potential U(x)')
plt.grid(True)
plt.show()
```

## Part (b) and (c): Equilibrium Points and Stability Analysis

Equilibrium points occur where the first derivative of the potential with respect to position is zero.


```python
x = symbols('x', real=True)

U = -W * (d**2 * (x**2 + 4*d**2)) / (x**4 + 128*d**4)
```

### First Derivative of the Potential


```python
dUdx = diff(U, x)
```

### Solve for Equilibrium Points


```python
equilibrium_points = solve(Eq(dUdx, 0), x)
print("Equilibrium points:", equilibrium_points)
```

### Second Derivative Test for Stability


```python
d2Udx2 = diff(dUdx, x)

for point in equilibrium_points:
    second = d2Udx2.subs(x, point)
    if second > 0:
        print(f"x = {point} is a STABLE equilibrium")
    elif second < 0:
        print(f"x = {point} is an UNSTABLE equilibrium")
    else:
        print(f"x = {point}: stability inconclusive")
```

## Part (d): Turning Points for a Given Total Energy

The turning points occur where the total energy equals the potential energy.


```python
E = -W / 32
```

### Plot of Potential with Energy Line


```python
plt.axhline(E, color='blue', linestyle='--', label='E = -W/32')
plt.plot(x_vals, U_vals)
plt.axhline(0, color='black', linewidth=0.5)
plt.xlabel('x')
plt.ylabel('U(x)')
plt.title('Potential U(x) with Energy Level')
plt.legend()
plt.grid(True)
plt.show()
```
